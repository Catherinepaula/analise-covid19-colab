```python
# 1. Contexto
from IPython.display import Markdown
Markdown("## 1. Contexto\n\nEste projeto de análise exploratória de dados (EDA) foca em um conjunto fictício de informações logísticas. Os dados representam pedidos de clientes, contendo região, categoria do produto, transportadora, datas de envio e entrega, e outras variáveis comerciais.\n\n**Objetivos:**\n- Investigar padrões de tempo de entrega.\n- Verificar diferenças entre transportadoras e regiões.\n- Identificar possíveis gargalos logísticos.\n- Fornecer insights para melhoria na operação de entregas.")

# 2. Pacotes e bibliotecas
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")

# 3. Exploração de dados
dados = """
pedido_id,cliente,regiao,produto,categoria,qtd,preco_unitario,data_envio,data_entrega,transportadora
1001,Ana,Centro,Camiseta,Vestuário,2,50.00,2023-01-10,2023-01-13,Transportex
1002,Bruno,Sul,Tênis,Calçados,1,150.00,2023-01-11,2023-01-16,LogRapid
1003,Carla,Norte,Notebook,Eletrônicos,1,2500.00,2023-01-12,2023-01-20,FastLog
1004,Daniela,Sudeste,Mochila,Acessórios,3,120.00,2023-01-13,2023-01-14,Transportex
1005,Evandro,Centro,Celular,Eletrônicos,1,1900.00,2023-01-14,2023-01-18,LogRapid
1006,Fabiana,Sul,Blusa,Vestuário,2,70.00,2023-01-15,2023-01-19,FastLog
"""
with open("logistica_ebac.csv", "w") as f:
    f.write(dados.strip())

df = pd.read_csv("logistica_ebac.csv", parse_dates=["data_envio", "data_entrega"])
df.head()

# 4. Manipulação
df["tempo_entrega"] = (df["data_entrega"] - df["data_envio"]).dt.days
df["categoria"] = df["categoria"].str.title()
df["valor_total"] = df["qtd"] * df["preco_unitario"]

# 5. Visualização

# Tempo médio de entrega por transportadora
plt.figure(figsize=(8,5))
sns.barplot(data=df, x="transportadora", y="tempo_entrega")
plt.title("Tempo Médio de Entrega por Transportadora")
plt.xlabel("Transportadora")
plt.ylabel("Tempo de Entrega (dias)")
plt.show()

# Valor total por categoria
plt.figure(figsize=(8,5))
sns.barplot(data=df, x="categoria", y="valor_total", estimator=sum)
plt.title("Valor Total Vendido por Categoria")
plt.xlabel("Categoria")
plt.ylabel("Valor Total")
plt.show()

# Distribuição do tempo de entrega
plt.figure(figsize=(6,4))
sns.histplot(df["tempo_entrega"], kde=True, bins=5)
plt.title("Distribuição dos Tempos de Entrega")
plt.xlabel("Dias")
plt.ylabel("Frequência")
plt.show()
```
